간단하게 matplotlib 및 seaborn 라이브러리를 이용하여 그래프를 그리는 방법을 알아보겠습니다.

matplotlib 및 seaborn 라이브러리를 사용해 본 적 있는 분들은 6번으로 이동하셔서 상관행렬 시각화부터 보시면 됩니다. 

```python
# 필요한 라이브러리 불러오기

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

# 밑의 두 줄은 맑은고딕(한글설정)을 불러오기 위한 코드입니다. 
# 그래프의 한글이 깨져서 나오는 경우 아래 코드 적어주면 됩니다. 
# 코드에서 에러날경우 마지막 두 줄 지우고 돌려주세요.

plt.rcParams['font.family'] = 'Malgun Gothic'       
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
```

위의 라이브러리에 대해 각각 간단히 설명하자면, 

- Numpy: 다차원 배열 및 수학 함수를 다루는 라이브러리

- Pandas: 테이블 형태의 데이터셋을 쉽게 다룰 수 있는 라이브러리

- Matplotlib: 다양한 그래프를 그릴 수 있는 기본적인 시각화 라이브러리

- Seaborn: Matplotlib 기반의 고급 그래프 라이브러리

### **1. 선그래프**

matplotlib.pyplot(이하 plt)를 통해 선그래프를 그리는 기본적인 코드는 다음과 같습니다. 

x축과 y축에 해당하는 데이터를 지정한 후, marker와 linestyle 등의 옵션을 변경할 수 있습니다. 

선그래프는 기본적으로 수치형 데이터의 시각화 방법입니다.

```python
# 선그래프 그리기
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='-')
plt.title("선그래프 연습")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### **2. 막대그래프(Bargraph)**

막대그래프는 기본적으로 범주형 데이터에 대해 그립니다. 

기본적으로는 x축에 해당하는 레이블(범주)를, y축에 빈도 데이터를 할당합니다. 그래프 색깔을 지정할 수 있습니다. 


```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color='powderblue')
plt.title("막대그래프 연습")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### **3. 히스토그램(Histogram)**

히스토그램은 수치형 데이터를 대상으로 그릴 수 있습니다. 

x축에 해당 데이터를, y축에는 빈도를 할당하고, 몇 개의 구간으로 나눌지를 bins를 통해 설정 가능합니다. 

그래프 색(color) 및 각 막대의 구분선(edgecolor)를 지정할수 있습니다. 

```python
data = np.random.randn(1000)  # 정규분포를 따르는 데이터를 생성하는 코드입니다. 

plt.hist(data, bins=30, color='darksalmon', alpha=0.7, edgecolor = 'white')
plt.title("히스토그램 연습")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### **4. 박스플롯(Boxplot)**

박스플롯은 수치형 데이터에 대해 반드시 확인해야 하는 중요한 데이터 시각화 방법 중 하나입니다. 이상치를 쉽게 확인할 수 있기 때문이죠.

seaborn 라이브러리를 사용하면 palette를 이용해 각 그래프에 색깔을 지정할 수 있습니다. 

```python
# 샘플 데이터 생성
data = [np.random.randn(100) for _ in range(4)]

sns.boxplot(data=data, palette='pastel')
plt.title("히스토그램 연습")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

1번부터 5번까지를 보셨다면 눈치 채셨을지도 모르겠지만,

사실 matplotlib이나 seaborn을 이용한 시각화 코드의 경우 다 비슷비슷하게 생겼습니다.

### **5. 산점도(Scatterplot)**

산점도의 경우 두 변수의 상관관계를 관찰하려는 목적으로 그리게 됩니다. 

x축에 첫 번쨰 변수를, y축에 두 번째 변수를 할당하고 산점도를 그려 그 양상을 확인하면 됩니다. 

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red')
plt.title("산점도 연습")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### **6. 상관행렬 (Correlation Matrix)**

지금까지 살펴본 그래프의 경우 하나의 변수 혹은 두 개의 변수에 대하여 시각화하고, 그 결과를 확인하는 그래프였습니다. 

그런데 데이터셋의 경우 보통은 하나의 타겟열과 여러 개의 변수(Features, 혹은 column)을 갖습니다. 

이러한 여러 변수들의 상관관계를 쉽게 관찰할 수 있는 것이 상관행렬(Correlation Matrix)입니다. 
<br><br>

지난 시간에 다루었던 Titanic train 데이터를 이용하여 상관행렬을 그리는 코드를 직접 써봅시다. 

```python
# 데이터셋 불러오기

df = pd.read_csv("C:/Users/UserPC/OneDrive/titanic_train.csv")  # 데이터셋을 불러옵니다. 

print(df.columns)   # 불러온 데이터셋의 column명 확인
df.head()   # 불러온 데이터셋의 첫 5개 데이터 확인
```

천천히 진행해 봅시다. 

위의 .head()를 확인하면, PassengerId열과 Name열, Ticket열의 경우 노이즈가 많은 데이터임을 확인할 수 있습니다. 

따라서 이 제 열들을 삭제해 줍니다. 

```python
# PassengerId, Name, Ticket열 삭제

df.drop(columns = ['PassengerId','Name','Ticket'], inplace=True)
df.head()
```

상관행렬은 기본적으로 숫자들 사이의 상관관계를 나타내 줍니다. 

상관행렬을 그리기 위해서는 object형 변수를 숫자형으로 바꾸어주어야 합니다. 
<br><br>

이를 위해 인코딩(Encoding)을 진행합니다. 인코딩(Encoding)이란 쉽게말해 문자를 숫자로 바꾸는 과정이며

여기서는 가장 쉬운 방법인 Label Encoding을 사용해 보겠습니다. 


```python
from sklearn.preprocessing import LabelEncoder

obj_cols = df.select_dtypes(include=['object']).columns     # df에서 object형 변수만 선택하여 obj_cols에 저장합니다.
le = LabelEncoder()

for col in obj_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()

# head()를 출력해보면 실제로 모든 데이터가 숫자로 바뀐 것을 확인할 수 있습니다. 
```

이제 마지막으로 상관행렬을 그려볼 차례입니다. 

상관행렬을 그리기 위해서는 우선 df.corr()를 사용해(Pandas에서 제공하는 함수) 상관계수 행렬을 구한 다음, 

이것을 seaborn의 heatmap을 이용해 그래프로 그려주면 됩니다. 
<br>

```python
corr_matrix = df.corr()

corr_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('상관행렬 연습')
plt.show()
```
